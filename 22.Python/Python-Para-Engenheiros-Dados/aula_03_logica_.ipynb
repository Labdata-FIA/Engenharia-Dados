## O problema

Dados do mundo real chegam sujos: campos faltando, valores negativos onde só pode haver positivo, datas mal formatadas. Seu código precisa **verificar antes de processar**.

## 1. Condições com `if / elif / else`

````python
# Classificando registros por volume
quantidade = 45_000

if quantidade == 0:
    classificacao = 'vazio'
elif quantidade < 1_000:
    classificacao = 'pequeno'
elif quantidade < 100_000:
    classificacao = 'médio'
else:
    classificacao = 'grande'

print(f'Lote de {quantidade} registros → {classificacao}')
````

**📝 O que essa célula faz:**

Usa `if / elif / else` para classificar a variável `quantidade` em uma categoria de texto, dependendo da faixa de valor.

```python
if quantidade == 0:
    classificacao = 'vazio'
elif quantidade < 1_000:
    classificacao = 'pequeno'
elif quantidade < 100_000:
    classificacao = 'médio'
else:
    classificacao = 'grande'
```

**🔀 Como o Python lê esse bloco:**

Ele testa as condições **em ordem**, de cima para baixo, e executa apenas o **primeiro** bloco cuja condição for verdadeira — ignorando todos os outros depois disso.

Com `quantidade = 45_000`: a primeira condição (`== 0`) é falsa, a segunda (`< 1_000`) também é falsa, mas a terceira (`< 100_000`) é **verdadeira** — então `classificacao` recebe `'médio'`, e o Python nem chega a olhar o `else`.

**`elif`** é a contração de "else if" ("senão, se") — permite testar várias condições em sequência. **`else`** é o "caso nenhuma das anteriores seja verdadeira".

## 2. Operadores de comparação

````python
tamanho = 500  # MB

print(tamanho > 100)    # maior que
print(tamanho < 1000)   # menor que
print(tamanho >= 500)   # maior ou igual
print(tamanho == 500)   # igual (dois sinais de =)
print(tamanho != 0)     # diferente
````

**📝 O que essa célula faz:**

Testa os principais **operadores de comparação** do Python, todos devolvendo `True` ou `False`.

| Operador | Significado | Exemplo |
|---|---|---|
| `>` | maior que | `tamanho > 100` |
| `<` | menor que | `tamanho < 1000` |
| `>=` | maior ou igual | `tamanho >= 500` |
| `==` | igual a | `tamanho == 500` |
| `!=` | diferente de | `tamanho != 0` |

**⚠️ Atenção ao detalhe mais comum de confundir:** `=` (um sinal) **atribui** um valor a uma variável (`x = 5` significa "x agora vale 5"). `==` (dois sinais) **compara** se dois valores são iguais (`x == 5` significa "x é igual a 5?", devolvendo `True` ou `False`). Esquecer um dos sinais de `=` é um dos erros mais comuns de quem está aprendendo.

## 3. Operadores lógicos: `and`, `or`, `not`

````python
status = 'sucesso'
registros = 15_000

# AND: as duas condições precisam ser verdadeiras
if status == 'sucesso' and registros > 0:
    print('Ingestão válida — pode promover para silver')

# OR: pelo menos uma precisa ser verdadeira
if status == 'falha' or registros == 0:
    print('Problema detectado — verificar origem')

# NOT: inverte
processado = False
if not processado:
    print('Ainda não processado — adicionar à fila')
````

**📝 O que essa célula faz:**

Combina condições usando os operadores lógicos `and`, `or` e `not`.

```python
if status == 'sucesso' and registros > 0:
    print('Ingestão válida...')
```

**🔗 `and` (E)** → a condição completa só é verdadeira se **as duas partes** forem verdadeiras. Aqui: o status precisa ser `'sucesso'` **E** a quantidade de registros precisa ser maior que zero. Se qualquer uma das duas for falsa, o bloco inteiro é falso.

**🔗 `or` (OU)** → a condição é verdadeira se **pelo menos uma** das partes for verdadeira. Aqui: "o status é falha **OU** os registros são zero" — se qualquer um dos dois acontecer, já é considerado um problema.

**🔄 `not` (NÃO)** → inverte o valor lógico. Se `processado` é `False`, então `not processado` vira `True`. É uma forma de ler "se NÃO está processado" de forma mais natural do que escrever `if processado == False`.

## 4. Validação de dados

O padrão mais comum em pipelines: verificar um registro antes de processá-lo.

````python
def validar_registro(registro):
   
    """Verifica se um registro tem os campos mínimos."""
    # Campo obrigatório ausente
    if registro.get('id') is None:
        return False, 'id ausente'
    
    # Valor inválido
    if registro.get('valor', 0) < 0:
        return False, 'valor negativo'
    
    # Formato esperado
    origem = registro.get('origem', '')
    if not origem or len(origem) < 3:
        return False, 'origem inválida'
    
    return True, 'ok'

# Testando
registros_teste = [
    {'id': 1, 'origem': 'api_crm',  'valor': 250.0},
    {'id': 2, 'origem': 'db',       'valor': -10.0},
    {'id': None, 'origem': 'erp',   'valor': 100.0},
    {'id': 4, 'origem': '',         'valor': 50.0},
]

for rec in registros_teste:
    valido, motivo = validar_registro(rec)
    icone = '✓' if valido else '✗'
    print(f"{icone} id={rec.get('id')} → {motivo}")
````

**📝 O que essa célula faz:**

Define uma **função** chamada `validar_registro` que recebe um dicionário e devolve se ele é válido — e por quê, caso não seja.

```python
def validar_registro(registro):
    if registro.get('id') is None:
        return False, 'id ausente'
    ...
    return True, 'ok'
```

**↩️ `return False, 'id ausente'` — duas coisas de uma vez?**

Sim! Quando você separa valores por vírgula depois do `return`, o Python os agrupa automaticamente em uma **tupla** — uma estrutura parecida com uma lista, mas que não pode ser alterada depois de criada. Isso permite que a função devolva **dois resultados ao mesmo tempo**: se é válido (`True`/`False`) e o motivo (texto explicando por quê).

**📦 Como usar esse retorno duplo (visto no código de teste abaixo):**
```python
valido, motivo = validar_registro(rec)
```
Essa linha "desempacota" a tupla retornada, colocando o primeiro valor em `valido` e o segundo em `motivo` — em uma única linha.

**🔍 A lógica da função, passo a passo:** ela testa três regras em sequência, e **assim que encontra um problema**, já devolve `False` com o motivo e **para ali** — não chega a checar as regras seguintes. Só devolve `True, 'ok'` no final, se passar por todas as checagens.

## 5. Tratando erros com `try / except`

Mesmo código correto pode encontrar situações inesperadas: arquivo não existe, rede caiu, dado no formato errado. O `try/except` evita que seu pipeline quebre completamente.

````python
# Sem try/except — qualquer problema para tudo
# valor_str = 'abc'
# valor = int(valor_str)  # ← isso quebraria o script

# Com try/except — trate o problema e continue
def converter_para_inteiro(texto):
    try:
        return int(texto)
    except ValueError:
        print(f'Aviso: "{texto}" não é um número válido — usando 0')
        return 0

print(converter_para_inteiro('1500'))   # funciona
print(converter_para_inteiro('abc'))    # erro tratado
print(converter_para_inteiro(''))       # erro tratado
````

**📝 O que essa célula faz:**

Mostra o problema que o `try/except` resolve, e depois a solução.

**❌ O comentário do início mostra o problema:**
```python
# valor_str = 'abc'
# valor = int(valor_str)  # ← isso quebraria o script
```
Tentar converter o texto `'abc'` para número inteiro com `int()` gera um erro (`ValueError`), porque `'abc'` não é um número. Sem tratamento, esse erro **para todo o programa** imediatamente.

**✅ A solução com `try/except`:**
```python
def converter_para_inteiro(texto):
    try:
        return int(texto)
    except ValueError:
        print(f'Aviso: "{texto}" não é um número válido — usando 0')
        return 0
```

**Como funciona:** o Python tenta executar o que está dentro do `try`. Se der erro, ele **não quebra o programa** — em vez disso, salta direto para o bloco `except` correspondente ao tipo de erro (aqui, `ValueError`, que é o erro de "valor incorreto para conversão"). Assim, em vez de travar, a função avisa o problema e devolve `0` como valor padrão.

**Resultado:** `converter_para_inteiro('1500')` funciona normalmente. `converter_para_inteiro('abc')` e `converter_para_inteiro('')` não quebram o programa — apenas avisam e devolvem `0`.

````python
# Exemplo mais próximo da realidade: processar um lote com erros
lote_bruto = [
    {'id': 1, 'quantidade': '1500'},
    {'id': 2, 'quantidade': 'N/A'},
    {'id': 3, 'quantidade': '8900'},
    {'id': 4, 'quantidade': None},
]

resultados = []
for item in lote_bruto:
    try:
        qtd = int(item['quantidade'])
        resultados.append({'id': item['id'], 'quantidade': qtd, 'valido': True})
    except (TypeError, ValueError):
        resultados.append({'id': item['id'], 'quantidade': 0, 'valido': False})

for r in resultados:
    marca = '✓' if r['valido'] else '✗'
    print(f"{marca} id={r['id']} quantidade={r['quantidade']}")
````

**📝 O que essa célula faz:**

Aplica o `try/except` dentro de um `for`, processando uma lista inteira de registros que tem alguns valores problemáticos.

```python
for item in lote_bruto:
    try:
        qtd = int(item['quantidade'])
        resultados.append({'id': item['id'], 'quantidade': qtd, 'valido': True})
    except (TypeError, ValueError):
        resultados.append({'id': item['id'], 'quantidade': 0, 'valido': False})
```

**🎯 Por que `except (TypeError, ValueError)` com dois tipos de erro entre parênteses?**

Porque os dados problemáticos podem falhar por **motivos diferentes**:
- `'N/A'` → não é um número → gera `ValueError`
- `None` → não é nem texto, é "nada" → gera `TypeError`

Colocando os dois tipos entre parênteses no mesmo `except`, o Python trata **qualquer um dos dois** da mesma forma. Se você só tivesse `except ValueError`, o caso do `None` quebraria o programa mesmo assim.

**Resultado:** o laço processa os 4 itens da lista sem travar, mesmo com `'N/A'` e `None` no meio — marcando cada um como válido (`True`) ou não (`False`).
