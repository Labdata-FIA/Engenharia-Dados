## O problema

Nas aulas anteriores, escrevemos código diretamente nas células. Mas imagine um pipeline que precisa limpar texto em 10 lugares diferentes — você copiaria o mesmo código 10 vezes?

Funções resolvem isso: **escreva uma vez, use em qualquer lugar**.

## 1. O que é uma função

Uma função tem:
- **Nome** — como você a chama
- **Parâmetros** — o que ela recebe
- **Corpo** — o que ela faz
- **Retorno** — o que ela devolve

````python
# Função mais simples possível
def saudar():
    print('Olá Fia!')

saudar()  # chamando a função
````

**📝 O que essa célula faz:**

Define a função mais simples possível, sem parâmetros e sem retorno, e depois a chama.

```python
def saudar():
    print('Olá Fia!')

saudar()
```

**🧱 As partes de uma função:**

- `def` → palavra-chave que diz "estou definindo uma função"
- `saudar` → o nome que você escolheu para a função
- `()` → parênteses vazios, porque essa função não recebe nenhum parâmetro
- `:` → indica que o que vem a seguir (indentado) é o **corpo** da função
- `print('Olá Fia!')` → o que a função executa quando é chamada

**⚠️ Importante:** definir a função (`def saudar(): ...`) não a executa! Ela só roda quando você a **chama** com `saudar()`. É como escrever uma receita de bolo — escrever a receita não faz o bolo aparecer; você precisa segui-la.

````python
# Função com parâmetro e retorno
def bytes_para_mb(bytes_valor):
    return bytes_valor / 1_048_576

tamanho = bytes_para_mb(524_288_000)
print(f'Tamanho: {tamanho:.2f} MB')
````

**📝 O que essa célula faz:**

Define uma função que **recebe um valor** (parâmetro) e **devolve um resultado** (retorno).

```python
def bytes_para_mb(bytes_valor):
    return bytes_valor / 1_048_576

tamanho = bytes_para_mb(524_288_000)
```

**📥 `bytes_valor` é o parâmetro** — um "espaço reservado" dentro da função que recebe o valor passado na hora da chamada. Quando você escreve `bytes_para_mb(524_288_000)`, o número `524_288_000` é colocado dentro de `bytes_valor`.

**📤 `return` devolve um valor** e **encerra a função imediatamente**. O resultado da conta `bytes_valor / 1_048_576` é devolvido para quem chamou a função, e fica guardado na variável `tamanho`.

**Diferença com a célula anterior:** a função `saudar()` usava `print()` (só mostra na tela, não guarda em nenhum lugar). Já `bytes_para_mb()` usa `return` (devolve um valor que pode ser guardado em uma variável e usado depois em outras contas). Essa é a diferença mais importante para entender sobre funções.

## 2. Funções de limpeza de dados

Transformações de dados são quase sempre funções: recebem um valor bruto, devolvem um valor limpo.

````python
# Limpar texto: remover espaços extras e padronizar capitalização
def limpar_texto(texto):
    if texto is None:
        return None
    return texto.strip().lower()

print(limpar_texto('  São Paulo  '))   # 'são paulo'
print(limpar_texto('BRASIL'))          # 'brasil'
print(limpar_texto(None))              # None
````

**📝 O que essa célula faz:**

Define uma função de limpeza de texto que lida com o caso de o valor ser `None`.

```python
def limpar_texto(texto):
    if texto is None:
        return None
    return texto.strip().lower()
```

**🛡️ Por que checar `if texto is None` primeiro?**

Porque se você tentasse chamar `.strip()` em um valor `None`, o Python geraria um erro (`None` não tem o método `.strip()`, só textos têm). Então a função primeiro checa: "esse valor é vazio? Se for, já devolvo `None` de volta e nem tento limpar." Só se o valor **não for `None`**, a segunda linha (`.strip().lower()`) é executada.

**🧹 `.strip()` e `.lower()` encadeados:**

- `.strip()` → remove espaços em branco do **início e do fim** do texto (mas não do meio). `'  São Paulo  '` se torna `'São Paulo'`
- `.lower()` → converte todo o texto para **minúsculas**. `'São Paulo'` se torna `'são paulo'`

Como os dois métodos são chamados **um atrás do outro** (`texto.strip().lower()`), o resultado do primeiro (`.strip()`) é imediatamente usado como entrada do segundo (`.lower()`) — sem precisar de uma variável intermediária.

````python
# Padronizar uma data de texto para formato ISO
def padronizar_data(data_texto):
    """Converte datas como '15/01/2024' para '2024-01-15'."""
    if data_texto is None:
        return None
    try:
        partes = data_texto.split('/')
        if len(partes) != 3:
            return None
        dia, mes, ano = partes
        return f'{ano}-{mes.zfill(2)}-{dia.zfill(2)}'
    except Exception:
        return None

print(padronizar_data('15/01/2024'))   # '2024-01-15'
print(padronizar_data('5/3/2024'))     # '2024-03-05'
print(padronizar_data('2024-01-15'))   # None (formato errado)
print(padronizar_data(None))           # None
````

**📝 O que essa célula faz:**

Converte uma data no formato brasileiro (`'15/01/2024'`) para o formato internacional ISO (`'2024-01-15'`), usado por bancos de dados e APIs.

```python
def padronizar_data(data_texto):
    try:
        partes = data_texto.split('/')
        if len(partes) != 3:
            return None
        dia, mes, ano = partes
        return f'{ano}-{mes.zfill(2)}-{dia.zfill(2)}'
    except Exception:
        return None
```

**✂️ `.split('/')`** → quebra o texto em pedaços, usando `/` como ponto de corte. `'15/01/2024'.split('/')` devolve a lista `['15', '01', '2024']`.

**📦 `dia, mes, ano = partes`** → "desempacota" a lista em três variáveis separadas de uma vez: a primeira posição vai para `dia`, a segunda para `mes`, a terceira para `ano`.

**0️⃣ `.zfill(2)`** → completa o texto com zeros à esquerda até ter 2 caracteres. `'5'.zfill(2)` vira `'05'`. Isso garante que `'5/3/2024'` (dia e mês com 1 dígito) também funcione corretamente, virando `'2024-03-05'` em vez de `'2024-3-5'`.

**🛡️ Por que `len(partes) != 3`?** Se a data não tiver exatamente 3 partes separadas por `/` (por exemplo, se já estiver em outro formato como `'2024-01-15'`, que não tem `/`), a função devolve `None` em vez de tentar continuar e gerar um erro.

**🚨 `except Exception`** → captura **qualquer** tipo de erro que possa acontecer dentro do `try` (mais genérico que `except ValueError`, por exemplo), garantindo que a função nunca quebre o programa, independente do problema.

````python
# Converter valor monetário de texto para número
def texto_para_valor(texto):
    """Converte 'R$ 1.250,50' para 1250.50."""
    if texto is None:
        return None
    try:
        limpo = texto.replace('R$', '').replace(' ', '')
        limpo = limpo.replace('.', '').replace(',', '.')
        return float(limpo)
    except (ValueError, AttributeError):
        return None

print(texto_para_valor('R$ 1.250,50'))   # 1250.5
print(texto_para_valor('R$ 300,00'))     # 300.0
print(texto_para_valor('abc'))           # None
````

**📝 O que essa célula faz:**

Converte um valor monetário escrito como texto brasileiro (`'R$ 1.250,50'`) para um número decimal Python (`1250.5`).

```python
def texto_para_valor(texto):
    try:
        limpo = texto.replace('R$', '').replace(' ', '')
        limpo = limpo.replace('.', '').replace(',', '.')
        return float(limpo)
    except (ValueError, AttributeError):
        return None
```

**🔄 `.replace(antigo, novo)`** → substitui um trecho de texto por outro. Aqui são usados **quatro `.replace()` encadeados**, um atrás do outro, cada um resolvendo um problema:

1. `.replace('R$', '')` → remove o símbolo de moeda → `'1.250,50'`
2. `.replace(' ', '')` → remove espaços → `'1.250,50'` (sem espaço entre R$ e o número)
3. `.replace('.', '')` → remove o ponto de **milhar** brasileiro → `'1250,50'`
4. `.replace(',', '.')` → troca a vírgula **decimal** brasileira pelo ponto decimal que o Python entende → `'1250.50'`

**🎯 Por que essa ordem específica?** Porque no Brasil usamos `.` para separar milhares e `,` para separar decimais (`1.250,50` = mil duzentos e cinquenta reais e cinquenta centavos), mas o Python (e a maioria dos sistemas internacionais) usa o contrário: `.` para decimais. Se você trocasse a ordem dos `.replace()`, o resultado ficaria errado.

**🚨 `except (ValueError, AttributeError)`** → `ValueError` cobre o caso de o texto não conseguir ser convertido em número (ex: `'abc'`). `AttributeError` cobre o caso de `texto` não ser nem um texto válido para começar (ex: receber `None`, que não tem o método `.replace()`).

## 3. Parâmetros com valor padrão

Funções podem ter parâmetros opcionais com um valor padrão:

````python
def montar_caminho(tabela, camada='bronze', formato='parquet'):
    return f's3://meu-lake/{camada}/{tabela}.{formato}'

# Usando os padrões
print(montar_caminho('clientes'))

# Sobrescrevendo parâmetros
print(montar_caminho('clientes', camada='silver'))
print(montar_caminho('relatorio', camada='gold', formato='csv'))
````

**📝 O que essa célula faz:**

Define uma função com **parâmetros que têm valor padrão** — ou seja, parâmetros opcionais.

```python
def montar_caminho(tabela, camada='bronze', formato='parquet'):
    return f's3://meu-lake/{camada}/{tabela}.{formato}'
```

**⚙️ O que significa `camada='bronze'` na definição da função?**

Significa: "se quem chamar essa função **não informar** um valor para `camada`, use `'bronze'` automaticamente." O mesmo vale para `formato='parquet'`.

**Três formas diferentes de chamar a mesma função:**
```python
montar_caminho('clientes')
# usa os padrões → camada='bronze', formato='parquet'

montar_caminho('clientes', camada='silver')
# sobrescreve só a camada, mantém formato='parquet'

montar_caminho('relatorio', camada='gold', formato='csv')
# sobrescreve os dois valores padrão
```

**💡 Por que isso é útil?** Porque na maioria das vezes os pipelines de um Data Lake escrevem na camada bronze em formato parquet — então definir esses como padrão evita repetir os mesmos argumentos toda vez, só sendo necessário especificá-los quando o caso for diferente do comum.

## 4. Funções que processam listas

Uma função pode receber uma lista inteira e devolver uma lista transformada:

````python
def limpar_lote(registros):
    """Aplica limpeza em uma lista de registros."""
    resultado = []
    for rec in registros:
        rec_limpo = {
            'id': rec.get('id'),
            'nome': limpar_texto(rec.get('nome')),
            'data': padronizar_data(rec.get('data')),
            'valor': texto_para_valor(rec.get('valor')),
        }
        resultado.append(rec_limpo)
    return resultado

# Testando com dados sujos
dados_brutos = [
    {'id': 1, 'nome': '  João Silva ', 'data': '10/01/2024', 'valor': 'R$ 1.500,00'},
    {'id': 2, 'nome': 'MARIA SOUZA',   'data': '5/2/2024',   'valor': 'R$ 250,50'},
    {'id': 3, 'nome': None,            'data': None,         'valor': None},
]

dados_limpos = limpar_lote(dados_brutos)
for rec in dados_limpos:
    print(rec)
````

**📝 O que essa célula faz:**

Cria uma função que recebe uma **lista inteira** de registros e devolve uma **nova lista** com cada registro já limpo — reaproveitando as funções `limpar_texto`, `padronizar_data` e `texto_para_valor` criadas antes.

```python
def limpar_lote(registros):
    resultado = []
    for rec in registros:
        rec_limpo = {
            'id': rec.get('id'),
            'nome': limpar_texto(rec.get('nome')),
            'data': padronizar_data(rec.get('data')),
            'valor': texto_para_valor(rec.get('valor')),
        }
        resultado.append(rec_limpo)
    return resultado
```

**🧩 Essa é a ideia central de "compor" funções:** em vez de escrever toda a lógica de limpeza de novo, a função `limpar_lote` simplesmente **chama** as três funções já criadas antes, uma para cada campo do registro. Isso é reutilização de código — você escreve a lógica de limpar texto/data/valor **uma única vez**, e usa em qualquer lugar que precisar.

**🔁 O laço `for rec in registros`** percorre cada dicionário da lista de entrada, cria um **novo** dicionário limpo (`rec_limpo`) e adiciona à lista `resultado`, que é devolvida no final. A lista original `registros` não é alterada — uma lista totalmente nova é construída.

## 5. Documentando funções com docstrings

Boa prática essencial em código de produção:

````python
def calcular_taxa_rejeicao(total, rejeitados):
    """
    Calcula a taxa de rejeição de um lote de dados.
    
    Parâmetros:
        total (int): total de registros do lote
        rejeitados (int): registros que não passaram na validação
    
    Retorna:
        float: taxa de rejeição entre 0.0 e 1.0
               Retorna 0.0 se total for zero.
    """
    if total == 0:
        return 0.0
    return rejeitados / total

taxa = calcular_taxa_rejeicao(1000, 37)
print(f'Taxa de rejeição: {taxa:.1%}')  # formato percentual

# Ver a documentação da função
help(calcular_taxa_rejeicao)
````

**📝 O que essa célula faz:**

Cria uma função com **docstring** (documentação) e usa a função `help()` para exibir essa documentação.

```python
def calcular_taxa_rejeicao(total, rejeitados):
    """
    Calcula a taxa de rejeição de um lote de dados.
    ...
    """
    if total == 0:
        return 0.0
    return rejeitados / total
```

**📖 O que é uma docstring?**

É um texto entre **três aspas** (`""" ... """`) colocado **imediatamente** depois da linha `def`. Serve para explicar o que a função faz, quais parâmetros recebe e o que devolve — é a documentação oficial da função, e o Python sabe reconhecê-la automaticamente.

**🆘 `help(calcular_taxa_rejeicao)`** → mostra na tela exatamente o conteúdo da docstring daquela função. Isso é extremamente útil quando você (ou um colega) esquece como uma função funciona — basta chamar `help()` nela em vez de abrir o código fonte.

**🛡️ `if total == 0: return 0.0`** → protege contra divisão por zero. Se `total` for `0`, dividir `rejeitados / total` geraria um erro (`ZeroDivisionError`). A função antecipa esse problema e devolve `0.0` diretamente nesse caso, sem nem tentar a divisão.

**`{taxa:.1%}`** → formata um número decimal como porcentagem com 1 casa decimal. `0.037` se torna `'3.7%'` — o `%` no formato já multiplica por 100 e adiciona o símbolo automaticamente.
